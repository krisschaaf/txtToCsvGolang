```python
import numpy as np
import matplotlib.pyplot as plt

data = np.genfromtxt('p2data.csv', delimiter=',', dtype=str)

dic = {}

for i, row in enumerate(data):
    if row[0] == '10.0.2.15':
        continue
    key = row[3]
    dic.setdefault(key, []).append(i)

averageDic = {}

for key in dic:
    indices = dic[key]
    values = [float(data[i][2]) for i in indices]
    averageDic[key] = np.mean(values)

print(averageDic)

index = np.linspace(1, len(dic.keys()), num=len(dic.keys()))

averageVals = []
keys = [key for key in averageDic]

for key in averageDic:
    averageVals.append(averageDic[key])

plt.bar(x=keys, height=averageVals, align='center')

plt.xticks(rotation='vertical')
plt.title("Durschnittliche Zeit der Hops pro IP Adresse")
plt.xlabel('IP Adresse')
plt.ylabel('Durschnittliche Zeit in Mikrosekunden')

plt.show()
```
